```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2d nonlinear function of x and y
def f(x, y):
    return np.sin(x) * np.cos(y) + 0.1 * (x**2 + y**2)
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# numerically compute hessian element f_xx, f_xy, f_yx, f_yy
eps = 1e-5
f_xx = (f(X+eps, Y) - 2*f(X, Y) + f(X-eps, Y)) / (eps**2)
f_yy = (f(X, Y+eps) - 2*f(X, Y) + f(X, Y-eps)) / (eps**2)
f_xy = (f(X+eps, Y+eps) - f(X+eps, Y-eps) - f(X-eps, Y+eps) + f(X-eps, Y-eps)) / (4*eps**2)
f_yx = f_xy  # mixed partials are equal
```

```python
# plot f(x,y) over a grid of x and y values, use viridis colormap, and add a colorbar

plt.imshow(Z, extent=[-2, 2, -2, 2], origin='lower', cmap='viridis')
plt.colorbar()
```
