```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Define your vivid side and your muted side
vivid_color = "#e74c3c" # Saturated Red
muted_color = "#95a5a6" # Muted Grey/Blue

# Create a diverging map: [Muted -> White -> Vivid]
colors = [muted_color, "white", vivid_color]
asym_cmap = mcolors.LinearSegmentedColormap.from_list("asym_div", colors)

my_cmap = mcolors.LinearSegmentedColormap.from_list("asym_div", ['#0092d1', '#eeeeee', '#8f7633'])

# Apply to data
plt.imshow(np.random.randn(10, 10), cmap=my_cmap)
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
val_list = [0.0, 0.7,0.8, 0.9]
for i in val_list:
    x = np.random.normal(size=10).cumsum()
    ax.plot(x, color=my_cmap(i), lw=4)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pypalettes import load_cmap, load_palette

cmap = load_cmap("BrowntoBlue_12")#, cmap_type="continuous")
palette = load_palette("BrowntoBlue_12")

data = np.random.randn(20, 20)

plt.imshow(data, cmap=cmap)
plt.colorbar()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
idx_list = [2, 3, 4, -2]
for i in idx_list:
    x = np.random.normal(size=10).cumsum()
    ax.plot(x, color=palette[i], lw=4)
```

```python
# Import CMasher to register colormaps
import cmasher as cmr

# Import packages for plotting
import matplotlib.pyplot as plt
import numpy as np

# Access rainforest colormap through CMasher or MPL
cmap = cmr.rainforest                   # CMasher
cmap = plt.get_cmap('cmr.iceburn')   # MPL

# Generate some data to plot
x = np.random.rand(100)
y = np.random.rand(100)
z = x**2+y**2

# Make scatter plot of data with colormap
plt.scatter(x, y, c=z, cmap=cmap, s=300)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
val_list = [0.2, 1.0, 0.95, 0.9]
for i in val_list:
    x = np.random.normal(size=10).cumsum()
    ax.plot(x, color=cmap(i), lw=4)
```
